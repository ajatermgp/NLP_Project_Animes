```python
import fandom
import io

fandom.set_wiki("detectiveconan")
```

```python

ges_dc = {}


while len(ges_dc) < 2159:
    ra_dc = fandom.random(500)
    ges_dc.update(ra_dc)
    
f_dc = io.open("dictionary_dc.txt", "a", encoding  ="utf-8")
for key in ges_dc:
    f_dc.write(key + "\n")
f_dc.close()

w_pointer_dc = io.open("dump_dc.txt", "a", encoding='utf-8')
file_dc = open("dictionary_dc.txt", encoding="utf-8")
for line in file_dc:
    if line != "":
        try:
            print(line)
            element_dc = fandom.page(line.rstrip("\n"))
            element_dc = element_dc.plain_text

            #print(element_dc)
            w_pointer_dc.write(element_dc + "\n\n")
        except:
            continue

w_pointer_dc.close()
```

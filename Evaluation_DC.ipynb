## Evaluation der Detective Conan Modelle

```python
import pandas as pd
from statistics import mean
from scipy.stats import f_oneway
```

```python
#angegeben werden die Semantic Textual Similarity Scores
#chatbot_only_dc_scores = unser Chatmodell mit den Detective Conan Daten
#chatgpt_scores = Antworten von ChatGPT 4.0 ohne zus√§tzliche Daten
#chatbot_all_animes_scores = unser Chatmodell mit den Detective Conan, One Piece und Naruto Daten

chatbot_only_dc_scores = [0.67,0.66,0.56,0.68,0.63,0.74,0.62,0.58,0.80,0.79,0.52,0.41,0.85,0.78,0.75]
chatgpt_scores = [0.70,0.74,0.63,0.78,0.71,0.78,0.72,0.79,0.77,0.71,0.62,0.50,0.77,0.60,0.68]
chatbot_all_animes_scores = [0.67,0.69,0.62,0.73,0.60,0.83,0.69,0.57,0.77,0.79,0.58,0.43,0.83,0.60,0.76]
```

```python
print(f"Mean Chatbot_only_DC Scores: {mean(chatbot_only_dc_scores)}")
print(f"Mean ChatGPT: {mean(chatgpt_scores)}")
print(f"Mean Chatbot_all_Animes Scores: {mean(chatbot_all_animes_scores)}")
```

```python
data = pd.DataFrame({
    'Modell': ['Chatmodell_DC-Daten'] * len(chatbot_only_dc_scores) + ['ChatGPT'] * len(chatgpt_scores) + ['Chatmodell_Anime-Daten'] * len(chatbot_all_animes_scores),
    'Similarity_Score': chatbot_only_dc_scores + chatgpt_scores + chatbot_all_animes_scores
})
data
```

```python
#Gibt es einen signifikanten Unterschied? 
#Da wir mehr als zwei Modelle miteinander vergleichen, nutzen wir ANOVA
anova_result = f_oneway(chatbot_only_dc_scores, chatgpt_scores, chatbot_all_animes_scores)
print(f"ANOVA Result: F={anova_result.statistic}, p={anova_result.pvalue}")
```

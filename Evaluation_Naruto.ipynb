## Evaluation der Naruto Modelle

```python
import pandas as pd
from statistics import mean
from scipy.stats import f_oneway
```

```python
#angegeben werden die Semantic Textual Similarity Scores
#chatbot_only_na_scores = unser Chatmodell mit den Naruto Daten
#chatgpt_scores = Antworten von ChatGPT 4.0 ohne zus√§tzliche Daten
#chatbot_all_animes_scores = unser Chatmodell mit den Naruto, One Piece und Detective Conan Daten

chatbot_only_na_scores = [0.33,0.53,0.71,0.78,0.73,0.66,0.74,0.76,0.70,0.72,0.75,0.58,0.61,0.80,0.69]
chatgpt_scores = [0.33,0.63,0.73,0.75,0.72,0.79,0.85,0.83,0.77,0.72,0.75,0.48,0.59,0.86,0.79]
chatbot_all_animes_scores = [0.40,0.58,0.68,0.75,0.73,0.70,0.66,0.83,0.70,0.72,0.73,0.53,0.62,0.78,0.84]
```

```python
print(f"Mean Chatbot_only_Na_Scores: {mean(chatbot_only_na_scores)}")
print(f"Mean ChatGPT: {mean(chatgpt_scores)}")
print(f"Mean Chatbot_all_Animes Scores: {mean(chatbot_all_animes_scores)}")
```

```python
data = pd.DataFrame({
    'Modell': ['Chatmodell_Na-Daten'] * len(chatbot_only_na_scores) + ['ChatGPT'] * len(chatgpt_scores) + ['Chatmodell_Anime-Daten'] * len(chatbot_all_animes_scores),
    'Similarity_Score': chatbot_only_na_scores + chatgpt_scores + chatbot_all_animes_scores
})
data
```

```python
#Gibt es einen signifikanten Unterschied? 
#Da wir mehr als zwei Modelle miteinander vergleichen, nutzen wir ANOVA
anova_result = f_oneway(chatbot_only_na_scores, chatgpt_scores, chatbot_all_animes_scores)
print(f"ANOVA Result: F={anova_result.statistic}, p={anova_result.pvalue}")
```

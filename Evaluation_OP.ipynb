## Evaluation der One Piece Modelle

```python
import pandas as pd
from statistics import mean
from scipy.stats import f_oneway
```

```python
#angegeben werden die Semantic Textual Similarity Scores
#chatbot_only_op_scores = unser Chatmodell mit den One Piece Daten
#chatgpt_scores = Antworten von ChatGPT 4.0 ohne zusätzliche Daten
#chatbot_all_animes_scores = unser Chatmodell mit den One Piece, Detective Conan und Naruto Daten

chatbot_only_op_scores = [0.68,0.59,0.86,0.80,0.71,0.45,0.71,0.76,0.75,0.75,0.67,0.70,0.42,0.62,0.74,0.84]
chatgpt_scores = [0.76,0.66,0.86,0.77,0.84,0.42,0.82,0.84,0.78,0.77,0.86,0.76,0.53,0.77,0.78,0.81]
chatbot_all_animes_scores = [0.77,0.57,0.88,0.79,0.84,0.34,0.78,0.79,0.79,0.78,0.74,0.82,0.54,0.61,0.75,0.78]
```

```python
print(f"Mean Chatbot_only_OP Scores: {mean(chatbot_only_op_scores)}")
print(f"Mean ChatGPT: {mean(chatgpt_scores)}")
print(f"Mean Chatbot_all_Animes Scores: {mean(chatbot_all_animes_scores)}")
```

```python
data = pd.DataFrame({
    'Modell': ['Chatmodell_OP-Daten'] * len(chatbot_only_op_scores) + ['ChatGPT'] * len(chatgpt_scores) + ['Chatmodell_Anime-Daten'] * len(chatbot_all_animes_scores),
    'Similarity_Score': chatbot_only_op_scores + chatgpt_scores + chatbot_all_animes_scores
})
data
```

```python
#Gibt es einen signifikanten Unterschied? 
#Da wir mehr als zwei Modelle miteinander vergleichen, nutzen wir ANOVA
anova_result = f_oneway(chatbot_only_op_scores, chatgpt_scores, chatbot_all_animes_scores)
print(f"ANOVA Result: F={anova_result.statistic}, p={anova_result.pvalue}")
```
